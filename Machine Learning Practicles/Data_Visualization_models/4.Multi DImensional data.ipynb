```python
#import 'numpy' as np for numerical operations
import numpy as np

#import 'matplotlib' for graphs and plots
import matplotlib.pyplot as plt
```

```python
#create data

#assigning number of data points
N=60

# x-values between 0.6 and 1.2 and y-values between 0 and 1.
g1=(0.6+0.6*np.random.rand(N),np.random.rand(N))

# x-values between 0.4 and 0.7 and y-values between 0 and 0.5.
g2=(0.4+0.3*np.random.rand(N),0.5*np.random.rand(N))

#x-values between 0 and 0.3 and y-values between 0 and 0.3.
g3=(0.3*np.random.rand(N),0.3*np.random.rand(N))

# x-values between 0.6 and 1.0 and y-values between 0 and 0.7.
g4 = (0.2 + 0.4 * np.random.rand(N), 0.7 * np.random.rand(N))
```

```python
#combine the groups
data=(g1,g2,g3,g4)

colors=("red","green","blue","purple")
groups=("coffee","tea","water","green tea")
```

## Creating the  scatter plot

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, facecolor="1.0")

for data, color, group in zip(data, colors, groups):
    x, y = data
    ax.scatter(x, y, alpha=0.8, c=color, edgecolors='none', s=30, label=group)

plt.title("Multi Scatter Plot")
plt.legend(loc=2)
plt.show()
```
