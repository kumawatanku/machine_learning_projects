```python
application_record=application_record.drop(columns=['OCCUPATION_TYPE','FLAG_MOBIL','FLAG_WORK_PHONE','FLAG_PHONE','FLAG_EMAIL'],axis=1)
print(application_record.info())
```

```python
credit_record.isnull().sum()
```

```python
df=pd.merge(application_record,credit_record,on='ID')
```

```python
df.head()
```

```python
df.shape
```

```python
def map_status(status):
    if status in ('C',0,'X'):
        return 1
    else:
        return 0
 
df['target']=df['STATUS'].apply(map_status)
```

```python
df=df.drop(columns='STATUS')
```

```python
df['CODE_GENDER']=df['CODE_GENDER'].map({'M':1,'F':0})
df['FLAG_OWN_CAR']=df['FLAG_OWN_CAR'].map({'Y':1,'N':0})
df['FLAG_OWN_REALTY']=df['FLAG_OWN_REALTY'].map({'Y':1,'N':0})
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['NAME_INCOME_TYPE']=le.fit_transform(df['NAME_INCOME_TYPE'])
df['NAME_EDUCATION_TYPE']=le.fit_transform(df['NAME_EDUCATION_TYPE'])
df['NAME_FAMILY_STATUS']=le.fit_transform(df['NAME_FAMILY_STATUS'])
df['NAME_HOUSING_TYPE']=le.fit_transform(df['NAME_HOUSING_TYPE'])
```

```python
df.info()
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
df=df.drop(columns='CNT_CHILDREN')
```

```python
df.info()
```

```python
df.tail()
```

```python
df['age']=abs(df['DAYS_BIRTH'])//365
df['years_employed']=abs(df['DAYS_EMPLOYED'])//365
```

```python
df=df.drop(columns=['DAYS_BIRTH','DAYS_EMPLOYED'],axis=1)
```

```python
df=df.drop(columns='ID')
```

```python
X=df.drop(columns='target')
y=df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print("Accuracy:",accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
df['target'].value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
classifier.fit(X_train,y_train)
y_predict=classifier.predict(X_test)
print("Accuracy:",accuracy_score(y_test,y_predict))
```
